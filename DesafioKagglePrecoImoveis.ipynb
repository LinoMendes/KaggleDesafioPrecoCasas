# Projeto para desafio do Kaggle - Preço dos Imóveis

```python
#Importação do Pandas
import pandas as pd
```

```python
#Importando o dataset de treino
base =pd.read_csv('train.csv')
base.head()
```

```python
#Visualizando o shape - as proporções da nossa base de dados 
Proporcoes = base.shape
print(f'A nossa base de dados tem Proporções de:{Proporcoes}' )
```

```python
# E as informações
base.info()
```

## Tratamento dos dados

```python
#Visualizando a quantidade de dados vazios
base.isnull()
```

```python
#Verificando em % de valores vazios por coluna
(base.isnull().sum()/base.shape[0]).sort_values(ascending=False).head(20)
```

```python
#Agora retirar as colunas que apresentam mais de 10% de valores vazios
excluir = base.columns[(base.isnull().sum()/base.shape[0]) > 0.1]
excluir
```

```python
base = base.drop(excluir, axis =1)
base.head()
```

- Vamos eleiminar **as colunas de texto**;
- Tratar **os valores vazios**;
- Escolher metodologias de análise para **validar o melhor modelo**

```python
# Primeiro eliminando colunas de texto
# Selecionando apenas as colunas numéricas
colunas = base.columns[base.dtypes != 'object']
colunas
```

```python
baseNumerica = base.loc[:, colunas]
baseNumerica.head()
```

```python
#Verificando valores vazios
baseNumerica.isnull().sum().sort_values(ascending=False).head()
```

```python
#Substituindo valores vazios por -1 - boa prática
baseNumerica = baseNumerica.fillna(-1)
baseNumerica.isnull().sum().sort_values(ascending=False).head()
```

## Criando e avaliando modelos

```python
#Selecionando X e Y
x = baseNumerica.drop('SalePrice',axis=1)
y = baseNumerica.SalePrice
```

```python
#Importação do train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Separando a base de treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state=42)
```

### Selecionando os algoritimos que vamos testar para otimizar o nosso resultado

```python
#Importando regressão linear
from sklearn.linear_model import LinearRegression
```

```python
#Criando a regressão linear e fazendo o fit com os dados de treino
reg_rl = LinearRegression().fit(x_train, y_train)
```

```python
#Fazendo a previsão para os dados de teste
y_rl = reg_rl.predict(x_test)
```

```python
#Importando a árvore de regressão
from sklearn import tree
```

```python
#Criando o regressor e fazendo o fit com os dados de treino
reg_ar = tree.DecisionTreeRegressor(random_state=42).fit(x_train, y_train)
```

```python
#Fazendo a previsão
y_ar = reg_ar.predict(x_test)
```

```python
#Importando o KNN
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Criando o regressor e fazendo o fit com os dados de treino
reg_knn = KNeighborsRegressor(n_neighbors=2).fit(x_train, y_train)
```

```python
#Fazendo a previsão
y_knn = reg_knn.predict(x_test)
```

### Avaliar os métodos utilizando o erro absoluto e o quadrático

```python
#Importando o erro médio absoluto
from sklearn.metrics import mean_absolute_error
#Importando o erro quadrático
from sklearn.metrics import mean_squared_error
```

```python
#Avaliando o erro na regressão 
print(mean_absolute_error(y_test, y_rl))
print(mean_squared_error(y_test, y_rl))
```

```python
#Avaliando o erro na árvor de decissão 
print(mean_absolute_error(y_test, y_ar))
print(mean_squared_error(y_test, y_ar))
```

```python
#Avaliando o erro no KNN 
print(mean_absolute_error(y_test, y_knn))
print(mean_squared_error(y_test, y_knn))
```
